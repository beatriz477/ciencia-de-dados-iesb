```python
# 1) Elabore um código que imprima a mensagem *"Olá terráqueos!!"* na tela.

print("Olá terráqueos")
```

```python
# 2) Elabore um código que peça um número e então mostre a mensagem: "O número informado foi [número]."

número = input("por favor, digite um número: ")
print(f"O número informado foi {número}")
```

```python
# 3) Elabore um código que peça dois números e imprima a soma.

número1 = int(input("Por favor, digite um número: "))
número2 = int(input("Por favor, digite outro número: "))
soma = número1 + número2
print(f"A soma dos números é {soma}")
```

```python
# 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.

nota1 = float(input("Por favor, digite a primeira nota: "))
nota2 = float(input("Por favor, digite a segunda nota: "))
nota3 = float(input("Por favor, digite a terceira nota: "))
nota4 = float(input("Por favor, digite a quarta nota: "))
Média = (nota1 + nota2 + nota3 + nota4) / 4
print(f"A média é {Média}")
```

```python
# 5) Elabore um código que converta metros para centímetros.

Metros = float(input("Por favor, digite a medida em metros: "))
Centímetros = Metros * 100
print(f"A medida convertida de metros para centímetros é de: {Centímetros}")
```

```python
# 6) Elabore um código que peça o raio de um círculo, calcule e mostre sua área

raio = float(input("Por gentileza, digite o raio do circulo: "))
Area = 3.14 * (raio ** 2)
print(f"A area do circulo é: {Area}")
```

```python
# 7) Elabore um código que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário

area = float(input("Por gentileza, digite a aréa do quadrado: "))
dobro = area * 2
print(f"O dobro da área do quadrado é: {dobro}")
```

```python
# 8) Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.

valor_hora = float(input("Por gentileza, digite o valor da hora trabalhada: "))
horas_trabalhadas = float(input("Por gentileza, digite a quantidade de horas trabalhadas: "))
Salario = valor_hora * horas_trabalhadas
print(f"você trabalhou {horas_trabalhadas} horas e ganhou R$ {Salario} no mês. Parabéns pelo seu imenso esforço!")
```

```python
# 9) Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. ** Onde C = 5 * ((F-32) / 9)**.

Fahrenheit = float(input("Por gentileza, digite a temperatura em fahrenheit: "))
Celsius = 5 * ((Fahrenheit - 32)) / 9
print(f"A temperatura em graus Celsius é: {Celsius}C")
```

```python
# 10) Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

Celsius = float(input("Por gentileza, digite a temperatura em celsius: "))
Fahrenheit = (Celsius * 9 / 5) + 32
print(f"A temperatura em graus fahrenheit é: {Fahrenheit}F")
```

```python
# 11) Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:
# A - o produto do dobro do primeiro com metade do segundo .
# B - a soma do triplo do primeiro com o terceiro.
# C - o terceiro elevado ao cubo.

numero1 = int(input("Por gentileza, digite um número inteiro: "))
numero2 = int(input("Por gentileza, digite outro número inteiro: "))
numero3 = float(input("Por gentileza, digite um número real: "))

A = (numero1 * 2) * (numero2 / 2)
B = (numero1 * 3) + numero3
C = numero3 ** 3

print(f"O produto do dobro de primeiro com a metade do segundo é: {A}")
print(f"A soma do triplo do primeiro com o terceiro é: {B}")
print(f"O terceiro número elevado ao cubo é: {C}")
```

```python
# 12) Elabore um código que receba os dados de entrada a altura de uma pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula: (72.7*altura) - 58

altura = float(input("Por favor, digite a sua altura: "))
peso_ideal = (72.7 * altura) - 58
print(f"seu peso ideal é de: {peso_ideal} kg")
```
